# VISUALIZING KMEANS WITH ELBOW METHOD

OPTIMUM NUMBER OF CLUSTERS K-means

```python
#Finding the optimum number of clusters for k-means classification
from sklearn.cluster import KMeans
WCSS =[]
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    WCSS.append(kmeans.inertia_)


#Plotting the results onto a line graph, allowing us to observe 'The elbow'
plt.plot(range(1,11),WCSS)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

 The elbow above drawn come to say the optimum cluster center of every columns where the elbow beds significantly

lets check with cluster center for each columns 

```python
kmeans.cluster_centers_
```

Cluster inertia Evaluation

```python
from sklearn.cluster import KMeans

wcss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters = i,init="k-means++",max_iter=300,n_init=10,random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_) # inertia is the sum of difference between centroid & corresponding data point
    print("k : ",i,"wcss : ",kmeans.inertia_)
```

```python
kmeans.score(x)
```
